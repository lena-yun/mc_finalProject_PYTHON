```python
import pandas as pd
from datetime import datetime

# 한글 폰트 문제 해결
import platform

from matplotlib import font_manager, rc
# plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~') 
```

```python
stock = pd.read_excel('./stock_final.xlsx')
```

```python
stock.head()
```

```python
stock.info()
```

```python
stock['kind'] = (stock['nongwoo']+stock['asiajongmyo'])/2
```

```python
stock['fertilizer'] = (stock['jobi']+stock['namhae'])/2
```

```python
stock['pesticide'] = (stock['kynong']+stock['dongbang'])/2
```

```python
stock['machine'] = (stock['dongyang']+stock['asiatech'])/2
```

```python
stock['smartfarm'] = (stock['greenplus']+stock['wooribio'])/2
```

```python
stock
```

```python
stock.info()
```

```python
stock_total = stock[['date','kind','fertilizer','pesticide','machine','smartfarm']]
```

```python
stock_total.info()
```

```python
cabbage = pd.read_excel('./stock_cabbage.xlsx')
```

```python
cabbage.info()
```

```python
cabbage
```

```python
# date 기준으로 left join
# result = pd.merge(cabbage, stock_total, on='date', how='outer')

result = pd.merge(cabbage, stock_total, left_on='date', right_on='date', how='left')
print(result)
```
