```python
# 한글 폰트 문제 해결
import platform

from matplotlib import font_manager, rc
# plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~') 
```

```python
webpage = requests.get('https://search.daum.net/search?w=tv&q=%EB%B0%B1%EC%A2%85%EC%9B%90%EC%9D%98%20%ED%91%B8%EB%93%9C%ED%8A%B8%EB%9F%AD%20%EC%8B%9C%EC%B2%AD%EB%A5%A0&irt=tv-program&irk=80982&DA=TVP')
soup = BeautifulSoup(webpage.content , 'html.parser') 
```

```python
soup
```

```python
table = soup.find('table', {'class' : 'tbl_rank'})
table
```

```python
rating_list = []
date_list = []

for trs in table.find_all('tr') :
    # print(trs)
    tds = list(trs.find_all('td'))
    # print(tds)
    for i, td in enumerate(tds) :
        if i == 2 :
            rating_list.append(td.text)
        if i == 0 :
            date_list.append(td.text)  
```

```python
print(rating_list)
print(date_list)
```

```python
import pandas as pd
```

```python
jibobbaek01_df = pd.DataFrame({
    'date' : date_list,
    'rating' : rating_list,
})
```

```python
df01 = jibobbaek01_df.loc[::-1]
df01
```

```python
df01 = df01.reset_index()
```

```python
df01.head()
```

```python
df01 = df01.drop(df01.columns[0], axis='columns')
```
