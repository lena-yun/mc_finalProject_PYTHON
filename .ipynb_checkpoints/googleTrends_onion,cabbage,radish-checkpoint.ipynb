```python
import pandas as pd
from datetime import datetime

# 한글 폰트 문제 해결
import platform

from matplotlib import font_manager, rc
# plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~') 
```

```python
holiday = pd.date_range(start='2013-01-01', end='2020-10-31')
```

```python
holiday_df = pd.DataFrame(holiday)
holiday_df.columns = ['date']
```

```python
holiday_df
```

```python
gt_onion_df = pd.read_csv('./gt_onion.csv', sep=',', names=['date','number'], header=None)
gt_cabbage_df = pd.read_csv('./gt_cabbage.csv', sep=',', names=['date','number'], header=None)
gt_radish_df = pd.read_csv('./gt_radish.csv', sep=',', names=['date','number'], header=None)
```

```python
gt_onion_df.info()
gt_cabbage_df.info()
gt_radish_df.info()
```

```python
gt_onion_df['date'] = pd.to_datetime(gt_onion_df['date'], format='%Y-%m-%d', errors='raise')
gt_onion_df.info()

gt_cabbage_df['date'] = pd.to_datetime(gt_cabbage_df['date'], format='%Y-%m-%d', errors='raise')
gt_cabbage_df.info()

gt_radish_df['date'] = pd.to_datetime(gt_radish_df['date'], format='%Y-%m-%d', errors='raise')
gt_radish_df.info()
```

```python
# onion
gt_merge_onion_df = pd.merge(holiday_df, gt_onion_df, how='left',on='date')
```

```python
gt_merge_onion_df.tail(20)
```

```python
gt_merge_onion_df.isnull().sum()
```

```python
gt_total_onion = gt_merge_onion_df.fillna(method='pad')
```

```python
gt_total_onion.head(10)
```

```python
gt_total_onion = gt_total_onion.fillna(22)
```

```python
gt_total_onion['product'] = '양파'
```

```python
gt_total_onion.head()
```

```python
# cabbage
gt_merge_cabbage_df = pd.merge(holiday_df, gt_cabbage_df, how='left',on='date')
gt_merge_cabbage_df.head(20)
```

```python
gt_merge_cabbage_df.isnull().sum()
```

```python
gt_total_cabbage = gt_merge_cabbage_df.fillna(method='pad')
```

```python
gt_total_cabbage.isnull().sum()
```

```python
gt_total_cabbage = gt_total_cabbage.fillna(25)
```

```python
gt_total_cabbage['product'] = '배추'
gt_total_cabbage.head()
```

```python
gt_total_cabbage.isnull().sum()
```

```python
# radish
gt_merge_radish_df = pd.merge(holiday_df, gt_radish_df, how='left',on='date')
gt_merge_radish_df.head(20)
```

```python
gt_merge_radish_df.isnull().sum()
```

```python
gt_total_radish = gt_merge_radish_df.fillna(method='pad')
```

```python
gt_total_radish.head(20)
```

```python
gt_total_radish = gt_total_radish.fillna(39)
```

```python
gt_total_radish['product'] = '무'
gt_total_radish.head()
```

```python
gt_total_radish.isnull().sum()
```

```python
# onion, cabbage, radish MERGE
googleTrends_onion_cabbage_radish = pd.concat([gt_total_onion, gt_total_cabbage, gt_total_radish], ignore_index=True) 
```

```python
googleTrends_onion_cabbage_radish.info()
```

```python
googleTrends_onion_cabbage_radish.head()
```

```python
googleTrends_onion_cabbage_radish.to_csv('googleTrends_onion_cabbage_radish.csv', mode='w', encoding='utf-8-sig', index=False) 

print('success')
```
